Create an `x`-array having values from 0 to 6 (stepsize = 0.1) and use that to calculate the values for $y = (x-3)^2 - 10$. Make a plot of that function and annotate the graphics with labels and title.

```python
x = arange(0, 6.1, 0.1)
f = lambda x : (x - 3)**2 - 10
y = f(x)

figure()
plot(x, y, 'r')
xlabel('$x$')
ylabel('$y$')
title('$(x-3)^2 - 10$')
grid()
show()
```

Use the same data to create the following 2D graphs:

a) a step-like graph   
b) a bar graph   
c) a scatterogram   
d) an area filled between the function and the x-axis

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row', figsize=(10, 10))
ax1.step(x, y)
ax2.bar(x, y)
ax3.scatter(x, y)
ax4.fill_between(x, y, 0)
```

Use the same data but make a plot, but now use

a) linewidth of 3   
b) dashed line type   
c) squared markers having size of 12 and color of red

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)
ax1.plot(x, y, linewidth=3)
ax2.plot(x, y, '--')
ax3.plot(x, y, 'rs', ms = 12);
```

Plot in the same graph the following approximations of log(x):

$f_1(x) = (x-1)$    
$f_2(x) = (x-1) - \frac{1}{2}(x-1)^2$   
$f_3(x) = (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3$   
$f_4(x) = (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3 - \frac{1}{4}(x-1)^4$   

Use the range $x \in [0, 2]$ with suitable stepsize. Add legends to the graphs to explain what the graphs are.

```python
x = linspace(0, 2, 1000)
f1 = lambda x : x - 1
f2 = lambda x : x - 1 - 1/2 * (x - 1)**2
f3 = lambda x : x - 1 - 1/2 * (x - 1)**2 + 1/3 * (x - 1)**3
f4 = lambda x : x - 1 - 1/2 * (x - 1)**2 + 1/3 * (x - 1)**3 - 1/4*(x - 1)**4

figure()
plot(x, f1(x), label='$f_1$')
plot(x, f2(x), label='$f_2$')
plot(x, f3(x), label='$f_3$')
plot(x, f4(x), label='$f_4$')
legend(loc= )
grid()
show()
```

Create two arrays ($x_1$ and $x_2$) of normally distributed random numbers ([`random.randn`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.randn.html)) both having 100 samples. Create a 2x2 subplots and show in them:

a) in upper left corner: [`scatter(x1, x2)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter)   
b) in upper right corner: [`hist(x1)`](http://matplotlib.org/api/pyplot_api.html?highlight=hist#matplotlib.pyplot.hist)   
c) in lower left corner: `hist(x2)`   
d) in lower right corner: `scatter(3*x1 + x2)`


```python
x1 = np.random.randn(100)
x2 = np.random.randn(100)

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (10, 10))
ax1.scatter(x1, x2)
ax2.hist(x1)
ax3.hist(x2)
ax4.scatter(3*x1, x2)
```

Create an array of normally distributed random numbers having a length of 10000. Make a histogram of the values. Count how many percent of the values are between -1.0 and 1.0 and show that value in the title of the graph.

```python
x = np.random.randn(10000)
percent = ((-1 < x) & (x < 1)).sum() / 10000 * 100
hist(x)
title(percent);
```
